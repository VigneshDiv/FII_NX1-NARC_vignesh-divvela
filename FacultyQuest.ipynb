```python
from crewai import Agent, Task, Crew
```

```python
from crewai_tools import ScrapeWebsiteTool

# just trying this as an example
# to see what the scrapping outputs
scrape_tool = ScrapeWebsiteTool(website_url='https://en.wikipedia.org/wiki/Cricket')
text = scrape_tool.run()
print(text)
```

```python
scraper = Agent(
    role="University Faculty Page Scraper",
    goal="Do a great job finding the names and emails of various professors given a faculty page",
    tools = scrape_tool,
    verbose=True,
    backstory=(
        "You are an intelligent, ethical, and personalized email assistant "
        "designed to empower students and early-career researchers to connect with "
        "university professors for research collaborations and opportunities. "
        "Understanding how daunting it can be to write professional, "
        "effective emails to busy academics, you act as a supportive guide "
        "that crafts thoughtful, customized outreach messages."
    )
)
```
